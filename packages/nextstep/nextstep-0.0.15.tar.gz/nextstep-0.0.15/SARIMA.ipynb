```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/Users/yuesongyang/Desktop/feature_mart_v3.csv')
train = data[-2879:-1].USEP
test = data[-2:-1].USEP
```

```python
plot_acf(train,lags = 20)
plt.show()
```

```python
plot_pacf(train, lags=60)
plt.show()
```

```python
# define model configuration
trend_AR = 1
trend_d = 1
trend_MA = 1

season_AR = 1
season_d = 0
season_MA = 1
season_period = 48

# define model
model = SARIMAX(train, order=(trend_AR, trend_d, trend_MA), 
                seasonal_order=(season_AR, season_d, season_MA, season_period))
model_fit = model.fit()
```

```python
yhat = model_fit.forecast()
yhat
```
